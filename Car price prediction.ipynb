# Selecting the Dependent and Independent Columns

```python
columns =["symboling", "wheelbase", "carlength", 
             "carwidth", "carheight", "curbweight", 
             "enginesize", "boreratio", "stroke", 
             "compressionratio", "horsepower", "peakrpm", 
             "citympg", "highwaympg"]
x = df[columns]
y = df['price']
print(x.shape)
print(y.shape)
```

# Train-Test-Split

```python
#Spliting the dataset into training and testing (80-20)

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
print('The shape of xtrain is:',xtrain.shape)
print('The shape of ytrain is:',ytrain.shape)
print('The shape of xtest is:',xtest.shape)
print('The shape of ytest is:',ytest.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(xtrain,ytrain)

#training set prediction
lr_predict = lr.predict(xtest)
lr_predict
```

```python
lr_score = round(lr.score(xtest,ytest) * 100, 2)

print("Accuracy is",lr_score)
```

# Decision Tree Regressor

```python
#Training the model and making predictions using the DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(xtrain, ytrain)
dtr_predict = dtr.predict(xtest)
print(dtr_predict)
```

```python
dtr_score = round(dtr.score(xtest,ytest) * 100, 2)

print("Accuracy is",dtr_score)
```

# Random Forest Regressor

```python
rfr = RandomForestRegressor()
rfr.fit(xtrain, ytrain)
rfr_predict = rfr.predict(xtest)
print(rfr_predict)
```

```python
rfr_score = round(rfr.score(xtest, ytest) * 100, 2)
print("Accuracy is:", rfr_score)
```

# Model Evaluation

```python
results = pd.DataFrame({'Models':['Linear Regression', 'Random Forest Regressor','Decision Tree Regressor'],
                      'Score':[lr_score,rfr_score,dtr_score]})
results
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x= results['Score'], y= results['Models'], orient='h')

plt.xlim(0,100)
plt.title('Accuracy of different models used:')
plt.xlabel('Accuracy Score')

plt.show()
```
